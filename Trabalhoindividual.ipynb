### **Gerar um relatório de progresso diário para mostrar o quão produtivo está sendo o trabalho dos funcionários e incluir no relatório outros itens como:**


---



* Total de Horas Trabalhadas
* Média Diária de Horas Trabalhadas
* Total de Bugs Corrigidos
* Média Diária de Bugs Corrigidos
* Total de Tarefas Concluídas
* Média Diária de Tarefas Concluídas
* Produtividade Diária (Tarefas Concluídas por Hora)


---



```python
import pandas as pd
import numpy as np
```

```python
pop_data = {'Dia': ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"],
            'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4 ],
            'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1 ],
            'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2 ]}

relatorio = pd.DataFrame(pop_data)
relatorio
```

```python
total_horas_trabalhadas = relatorio['Horas Trabalhadas'].sum()
media_diaria_horas = relatorio['Horas Trabalhadas'].mean()
total_bugs_corrigidos = relatorio['Bugs Corrigidos'].sum()
media_diaria_bugs = relatorio['Bugs Corrigidos'].mean()
total_tarefas_concluidas = relatorio['Tarefas Concluídas'].sum()
media_diaria_tarefas = relatorio['Tarefas Concluídas'].mean()

# Exibindo as métricas
print(f'Total de Horas Trabalhadas: {total_horas_trabalhadas} horas')
print(f'Média Diária de Horas Trabalhadas: {media_diaria_horas:.2f} horas por dia')
print(f'Total de Bugs Corrigidos: {total_bugs_corrigidos} bugs')
print(f'Média Diária de Bugs Corrigidos: {media_diaria_bugs:.2f} bugs por dia')
print(f'Total de Tarefas Concluídas: {total_tarefas_concluidas} tarefas')
print(f'Média Diária de Tarefas Concluídas: {media_diaria_tarefas:.2f} tarefas por dia')
```

```python
# Exibindo o DataFrame atualizado com a produtividade diária
print('\nRelatório com Produtividade Diária:')

print(relatorio[['Dia', 'Tarefas Concluídas', 'Horas Trabalhadas', 'Produtividade Diária']])
```
